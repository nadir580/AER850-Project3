```python
#Step2_YOLOv8_Training
!pip install ultralytics

from ultralytics import YOLO

from google.colab import drive
drive.mount('/content/drive')

yolo_model = YOLO('yolov8n.pt')

yolo_model.train(
    data="/content/drive/MyDrive/AER850/Project 3 Data/data/data.yaml",
    epochs=100,
    batch=6,
    imgsz=920,
    name="PCB_Component_Detection_Model"
)

yolo_model.save('/content/drive/MyDrive/AER850/Project 3 Data/PCB_Component_Detection_Model_trained.pt')
```

```python
# Step 3: YOLOv8 Evaluation

from ultralytics import YOLO

model = YOLO('/content/drive/MyDrive/AER850/Project 3 Data/runs/detect/PCB_Component_Detection_Model/weights/best.pt')

mega_ard = "/content/drive/MyDrive/AER850 Project 3 Files/data/evaluation/ardmega.jpg"
uno_ard = "/content/drive/MyDrive/AER850 Project 3 Files/data/evaluation/arduno.jpg"
raspi = "/content/drive/MyDrive/AER850 Project 3 Files/data/evaluation/rasppi.jpg"

model.predict(source=mega_ard, save=True, imgsz=928, conf=0.25, line_width=3)
model.predict(source=uno_ard, save=True, imgsz=928, conf=0.25, line_width=3)
model.predict(source=raspi, save=True, imgsz=928, conf=0.25, line_width=3)
```
