```python
#Step2_YOLOv8_Training
!pip install ultralytics

from ultralytics import YOLO

from google.colab import drive
drive.mount('/content/drive')

yolo_model = YOLO('yolov8n.pt')

yolo_model.train(
    data="/content/drive/MyDrive/AER850/Project 3 Data/data/data.yaml",
    epochs=100,
    batch=6,
    imgsz=960,
    name="PCB_Component_Detection_Model"
)

yolo_model.save('/content/drive/MyDrive/AER850/Project 3 Data/PCB_Component_Detection_Model_trained.pt')
```

```python
# Step 3: YOLOv8 Evaluation

# Define evaluation images
evaluation_images = [
    "/content/drive/MyDrive/AER850/Project 3 Data/data/evaluation/ardmega.jpg",
    "/content/drive/MyDrive/AER850/Project 3 Data/data/evaluation/arduno.jpg",
    "/content/drive/MyDrive/AER850/Project 3 Data/data/evaluation/rasppi.jpg"
]

# Load the trained model
from ultralytics import YOLO
yolo_model = YOLO('/content/drive/MyDrive/AER850/Project 3 Data/PCB_Component_Detection_Model_trained.pt')

# Run predictions and visualize results
import cv2
import matplotlib.pyplot as plt

for i, imgpath in enumerate(evaluation_images):

    # Prediction
    results = yolo_model.predict(imgpath, imgsz=900, conf=0.1)  # Adjust confidence as needed

    # Visualize results
    result_image = results[0].plot()

    # Save and display the result
    output_path = f"/content/drive/MyDrive/AER850/Project 3 Data/evaluation_results/evaluation_result_{i+1}.jpg"
    cv2.imwrite(output_path, result_image)
    print(f"Evaluation result saved: {output_path}")

    # Show the result
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
    plt.title(f"Evaluation Result {i+1}")
    plt.axis("off")
    plt.show()
```
